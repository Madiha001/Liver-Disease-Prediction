```python
#Import Python Libraries
import numpy as np
# import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
# import statsmodels.api as sm
from sklearn import preprocessing
```

```python
dat = pd.read_csv("Indian Liver Patient Dataset (ILPD).csv")
dat
```

```python
lablEncodr = preprocessing.LabelEncoder()
dataset = dat
dataset['Gender']=lablEncodr.fit_transform(dataset['Gender'])
ds=dataset.dropna() 
y = ds["Class"].values
X = ds.iloc[: , :10]
X
```

```python
#Results Before Preprocessing of Data
#Applying KNN classifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test,y_pred)
print('Confusion Matrix : \n', cm)
print("Accuracy of KNN for 25-75%: {0:.2}".format(metrics.accuracy_score(y_test, y_pred)))

total=sum(sum(cm))
sensitivity = cm[0,0]/(cm[0,0]+cm[0,1])
print('Sensitivity :{0:.2}'.format(sensitivity))

specificity = cm[1,1]/(cm[1,0]+cm[1,1])
print('Specificity : {0:.2}'.format(specificity))

print('MCC: {0:.2}'.format(metrics.matthews_corrcoef(y_test,y_pred)))
print("F-Measure: {0:.2}".format(metrics.f1_score(y_test,y_pred)))
print('AUROC: {0:.2}'.format(metrics.roc_auc_score(y_test, y_pred)))
```

```python
def outlier_identification(Col):
    lower_range = []
    upper_range = []
    sorted(Col)
    Q1, Q3 = np.percentile(Col , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
#finding min max values in Age column
lower,upper = outlier_identification(dataset['Age'])
print(lower)
print(upper)
#printing outliers
dataset[(dataset['Age']< lower) | (dataset['Age'] > upper)]
```

```python
#finding min max values in Age column
lower,upper = outlier_identification(dataset['Total Bilirubin'])
print(lower)
print(upper)
#printing outliers
dataset[(dataset['Total Bilirubin']< lower) | (dataset['Total Bilirubin'] > upper)]
```

```python
#Feature Extraction
def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in dataset.columns:
                    del dataset[colname] # deleting the column from the dataset

    print(dataset)
correlation(df,0.7)
```
